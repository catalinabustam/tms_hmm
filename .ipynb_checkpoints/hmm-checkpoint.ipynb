### Function to load data

```python
def remove_rois(rois, series):
    series_d = np.delete(series, rois, axis=1)
    return  series_d
```

```python
def get_time_series(dirpath, subects, stimulus, scan, parcel, scaler,rois ):

    subject_list = []
    stimuli_list =  []
    
    for subject in subjects:
        file_path = f'{dirpath}{subject}/{stimulus}/roi_timeseries_0/{scan}/_compcor_ncomponents_5_selector_pc10.linear1.wm1.global0.motion1.quadratic1.gm0.compcor1.csf1/_bandpass_freqs_0.01.0.1/{parcel}/clean_roi_csv/roi_stats.csv'
        series = np.genfromtxt(file_path,delimiter='\t',skip_header=1)
        series_w_z = remove_rois(rois, series)
        series_s = scaler.fit_transform(series_w_z)
        subject_list.append(series_s)
        stimuli_list.append(scan + stimulus)
    return np.array(subject_list), stimuli_list
```

### Function to create left to right transmat


```python
def create_lf_matrix(states,delta=None):
    transmat = np.triu(np.ones((states,states)))
    nonzero = np.nonzero(transmat)
  
    if delta:
        for i in range(len(nonzero[0])):
            m = nonzero[0][i]
            n = nonzero[1][i]
    
        if (n > (m+ delta)):
            transmat[m,n] = 0
  
    transmat = (transmat.T/np.count_nonzero(transmat, axis=1))
  
    return transmat.T
```

### Load and concatenate time series

```python
def get_all_data(scans, stimuli,scaler, rois ):
    all_c = np.empty((0, 600, 364))
    stimuli_list = []
    for scan in scans:
        for stimulus in stimuli:
            stimu_array, stimuli_l = get_time_series(dirpath, subjects, stimulus, scan, parcel_path, scaler, rois)
            all_c = np.append(all_c, stimu_array, axis=0)
            stimuli_list= stimuli_list + stimuli_l
    return all_c, stimuli_list
```

### Find the best model

```python
max_states = 17
cov_matrices = ['lr', 'ergodic']
max_gaussians = 6
```

1e-3

```python
hmm_summary = pd.DataFrame(columns={'states', 'cov_matrix', 'num_gauss', 'log_prob'})

for states in tqdm(range(3,max_states+1)):
    for cov_matrix in cov_matrices:
        for num_gauss in tqdm(range(1,max_gaussians+1)):
            if cov_matrix == 'lr':
                
                hmm_m = hmm.GMMHMM(n_components=states, n_mix = num_gauss, covariance_type="full", init_params="cm",n_iter=100)
                
                startprob = np.zeros(states)
                startprob[0] = 1
                hmm_m.startprob_ = startprob
                hmm_m.transmat_ = create_lf_matrix(states)
            else:
                hmm_m = hmm.GMMHMM(n_components=states, n_mix = num_gauss, covariance_type="full",n_iter=100)

            hmm_m.fit(all_data, lenghts)
            
            score = hmm_m.score(all_data, lenghts)
            
            
            hmm_model = pd.DataFrame([{'states': states, 'cov_matrix': cov_matrix, 'num_gauss': num_gauss , 'log_prob': score}]) 
            hmm_summary = pd.concat([hmm_summary,  hmm_model], ignore_index = True)            
            
```

```python
 def create_lf_matrix_w(states):
  transmat = np.zeros((states,states))

  for i in range(states):
    transmat[i,i:states] = 1/(states-i)
  return transmat
```

```python
def get_time_series_average(dirpath, subects, stimulus, scan, parcel, scaler,rois, networks_list):

    subject_list = []
    stimuli_list =  []
    
    for subject in subjects:
        file_path = f'{dirpath}{subject}/{stimulus}/roi_timeseries_0/{scan}/_compcor_ncomponents_5_selector_pc10.linear1.wm1.global0.motion1.quadratic1.gm0.compcor1.csf1/_bandpass_freqs_0.01.0.1/{parcel}/clean_roi_csv/roi_stats.csv'
        series = np.genfromtxt(file_path,delimiter='\t',skip_header=1)
        series_w_z = remove_rois(rois, series)
        networks_w_z =  np.delete(np.array(networks_list), rois)
        uniqs, idx = np.unique(networks_w_z, return_index=True)
        idx.sort()
        
        result = np.empty((series_w_z.shape[0], ))

        for i in range(len(uniqs)-1):
            val = idx[i]
            start, end = idx[i], idx[i+1]
            network_s = np.mean(series_w_z[:,start:end], axis=1)
            print(result.shape)
            print(network_s.shape)
            result = np.concatenate((result, network_s))
            
        print(result.shape)
        series_s = scaler.fit_transform(series_w_z)
        subject_list.append(series_s)
        stimuli_list.append(scan + stimulus)
    return np.array(subject_list), stimuli_list
```

```python
def get_all_data_average(scans, stimuli,scaler, rois, networks_list ):
    all_c = np.empty((0, 600, 364))
    stimuli_list = []
    for scan in scans:
        for stimulus in stimuli:
            stimu_array, stimuli_l = get_time_series_average(dirpath, subjects, stimulus, scan, parcel_path, scaler, rois, networks_list)
            all_c = np.append(all_c, stimu_array, axis=0)
            stimuli_list= stimuli_list + stimuli_l
    return all_c, stimuli_list
```
